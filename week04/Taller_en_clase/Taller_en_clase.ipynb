```python
import os
import re
import collections
import pandas as pd
from sklearn.metrics.pairwise import cosine_distances
import numpy as np
```

Carga del corpus en memoria

```python
# Define the path to the directory containing the text files
CORPUS_DIR = "data"
documents = {}
```

```python
print(os.listdir(CORPUS_DIR))
```

```python
def clean_text(text):
    # Eliminar caracteres no deseados
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text

for filename in os.listdir(CORPUS_DIR):
    if filename.endswith('.txt'):
        file_path = os.path.join(CORPUS_DIR, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read().lower()
            cleaned_text = clean_text(text)
            documents[filename] = cleaned_text
```

```python
print(documents)
```

```python
# Dictionary to save the doc normalized
normalized_word_counts = {}
```

```python
for doc in documents:
    word_count = collections.Counter(documents[doc].split())
    total_words = sum(word_count.values())
    normalized_word_count = {word: count/total_words for word, count in word_count.items()}
    normalized_word_counts[doc] = normalized_word_count
```

```python
df = pd.DataFrame.from_dict(normalized_word_counts)
```

```python
df_no_nan = df.fillna(0)
df_no_nan = df_no_nan.rename_axis('Files')
columnas = df_no_nan.columns.values[0:]
df1 = pd.pivot_table(df_no_nan, columns='Files', values=columnas)
df1
```

```python
df1.to_csv("vect_normalized.csv",index=True)
```

```python
columns_names = df1.columns.values
```

```python
df2 = pd.DataFrame(columns=columns_names)
df2
```

```python
with open('query.txt', 'r') as file:
    lineas = [line.rstrip('\n') for line in file]
df_querys = pd.DataFrame(lineas,columns = ["Consulta"])
df_querys['Ranking'] = pd.Series(dtype=str)
df_querys
```

```python
for linea in lineas:
    query = linea.split()
    vector = np.zeros(df2.shape[1], dtype=int)
    for palabra in query:
        if palabra in df2.columns:
            i = df2.columns.get_loc(palabra)
            vector[i] = 1
    df2.loc[len(df2)] = vector
    
df2
```

```python
df2.to_csv("querys_dictionary.csv",index=True)
```

```python
list_top = []

for idx_df1, fila_df1 in df2.iterrows():
    distancias_query = []
    for idx_df2, fila_df2 in df1.iterrows():
        distancia_coseno = cosine_distances([fila_df1.values], [fila_df2.values])[0][0]
        distancias_query.append((distancia_coseno, idx_df2))
    ranking = sorted(distancias_query, key=lambda x: x[0])
    top = ranking[:3]
    list_top.append((idx_df1, top))
```

```python
print(list_top)
```

```python
df_querys['Ranking'] = list_top
df_querys
```

```python
df_querys.to_csv("querys_ranked.csv",index=True)
```
